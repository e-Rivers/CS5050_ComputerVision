# Memory and Color

Import the needed libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

Load the photo into open CV
(It will be load as a Numpy array with dimensions: (height, width, rgb))

```python
img = cv.imread("mclovin.jpg")
```

```python
img.shape
```

Split the input image 'img' into its three color channels: blue (b), green (g), and red (r). Then, Merge the color channels back together, but in a different order: red, green, blue (r, g, b). This  converts the image from BGR color space (used by OpenCV) to RGB color space (commonly used in image display).


```python
b,g,r = cv.split(img)
img = cv.merge((r,g,b))
```

Show the picture

```python
plt.imshow(img)
plt.title("McLovin - Original")
plt.show()
```

## Change the Red and Green Channels

Split the input image into its three color channels: red (r), green (g), and blue (b). Then, merge the color channels back together buut swap the red and green channels.


```python
r,g,b = cv.split(img)
imgChannelChange = cv.merge((g,r,b))
```

Display the image with the red and green channels swapped.

```python
plt.imshow(imgChannelChange)
plt.title("McLovin - Changed R and G channels")
plt.show()
```

## Convert to Gray Scale

Convert the image to grayscale.

```python
imgGrayScale = cv.cvtColor(imgChannelChange, cv.COLOR_BGR2GRAY)
```

Display the image in grayscale.

```python
plt.imshow(imgGrayScale, cmap="gray")
plt.title("McLovin - Gray Scale")
plt.show()
```

## Invert Color

```python
imgInverted = 255 - img
```

```python
plt.imshow(imgInverted, cmap="gray")
plt.title("McLovin - Inverted")
plt.show()
```

## Convert the original image into HSV

Convert the original image 'img' from BGR color space to HSV (Hue, Saturation, Value).

```python
imgHSV = cv.cvtColor(img, cv.COLOR_BGR2HSV)
```

Split it into its three channels: hue (h), saturation (s), and value (v)

```python
h,s,v = cv.split(imgHSV)
```

Display the HSV

```python
plt.imshow(h)
plt.title("McLovin - HUE only")
plt.show()
```
